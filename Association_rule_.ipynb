```
transactions = []
for i in range(0, len(data)):
  transactions.append([str(data.values[i, j]) for j in range(0, 20) if str(data.values[i,j])!='0'])
```

```
!pip install apyori
```

```
from apyori import apriori
rules = apriori(transactions=transactions,min_support = 0.003,min_confidence=0.2,min_lift=3,min_length=2,max_length=2)
result = list(rules)
```

```
result
```

```
import pandas as pd
data = []
for r in result:
    for s in r.ordered_statistics:
        data.append([
            list(s.items_base)[0],
            list(s.items_add)[0],
            r.support,
            s.confidence,
            s.lift
        ])

df = pd.DataFrame(
    data,
    columns=["Base", "Add", "Support", "Confidence", "Lift"]
)

print(df)
```

```
print("{:<20} {:<20} {:<10} {:<12} {:<8}".format(
    "Base", "Add", "Support", "Confidence", "Lift"
))

for i in range(len(result)):
    for j in range(len(result[i].ordered_statistics)):
        base = list(result[i].ordered_statistics[j].items_base)[0]
        add = list(result[i].ordered_statistics[j].items_add)[0]
        support = result[i].support
        confidence = result[i].ordered_statistics[j].confidence
        lift = result[i].ordered_statistics[j].lift

        print("{:<20} {:<20} {:<10.4f} {:<12.4f} {:<8.2f}".format(
            base, add, support, confidence, lift
        ))
```
